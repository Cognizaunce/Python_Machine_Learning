```python
#2 Reading the Dataset
import pandas as pd
df  = pd.read_csv("avocado.csv")
print(len(df))
df.dropna()
```

```python
#3 Feature Extraction
df.drop(columns=['Unnamed: 0','Date', 'region'], inplace = True)# overwrites parent object
df #Note: can delete labeless labels with their error name
```

```python
# 3a encoding
from sklearn.preprocessing import LabelEncoder
df['type'] = LabelEncoder().fit_transform(df['type'])
df
```

```python
#3b Assign input attributes and outputs
features = len(df.columns)
y = df.iloc[:, 0].values

X = df.drop('AveragePrice', axis=1).values
```

```python
#3c Scale the input attributes
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
```

```python
#4 Split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, train_size= 0.8)
X_validate, X_test, y_validate, y_test = train_test_split(X_normalized, y, test_size= 0.5) # ignore order, validate should always be used last
```

```python
#5 Train model with sample KNN regressor
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor(n_neighbors = 8).fit(X_train, y_train)

model.score(X_test, y_test)
```

```python
#6a Find the best 'k' value
scores = []
results = 0
best_score = 0
neighbors = range(1,15)
for i in neighbors:
    knn = KNeighborsRegressor(n_neighbors=i).fit(X_train, y_train)
    results = knn.score(X_test, y_test)
    scores.append(round(results,2))

    if results > best_score:
        best_score = results
        best_k = i
        bestmodel = knn

print (scores)
print(best_k)
```

```python
#6b confirm using elbow plot
import matplotlib.pyplot as plt
plt.plot(neighbors, scores)
```

```python
#7 Get the final R-squared score of your final KNN regresor
accuracy = bestmodel.score(X_validate,y_validate)
print("the best model has an accuracy of:", round(accuracy,2))
```

```python
#8a use Linear Regression (sample of an alternative modelling agent)
# DO NOT DO THOROUGH TESTING!
from sklearn.linear_model import LinearRegression

modelLR = LinearRegression().fit(X_train, y_train)
score = modelLR.score(X_test, y_test)
print(score)
# The purpose of this exercise was to determine if it was viable to use.
# Conclusion: not worth implementing for this dataset
```

```python
#8b use Linear Regression (sample of an alternative modelling agent)
# DO NOT DO THOROUGH TESTING!
from sklearn.ensemble import RandomForestRegressor

modelRF = RandomForestRegressor(criterion='friedman_mse', max_leaf_nodes=100).fit(X_train, y_train)
#added max leaf nodes just in case model converges
score = modelRF.score(X_test, y_test)
print(score)
# The purpose of this exercise was to determine if it was viable to use.
# Conclusion: could worth be implementing for this dataset (send that in a memo for a report proposal)
```

```python
#9 print any outstanding analytics.
from sklearn.metrics import mean_squared_error

y_pred = modelRF.predict(X_test) #sample analytic

print("Accuracy:",score)
print("MSE" + str(mean_squared_error(y_test,y_pred)))
```
