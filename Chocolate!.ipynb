```python
df['CocoaPercent'] = df['CocoaPercent'].astype(str).str.replace(r'\D',"", regex=True).astype(int)
```

```python
df
```

```python
ratings = df.Rating.values
percent = df.CocoaPercent.values

plt.figure(figsize = (16,5))
plt.scatter(percent,  ratings, color = "b")
plt.xlabel("Cocoa Percent")
```

```python
df.describe()
```

```python
df.drop(df[df.CocoaPercent > 100].index, inplace = True)
```

```python
ratings = df.Rating.values
percent = df.CocoaPercent.values

plt.figure(figsize = (16,5))
plt.scatter(percent,  ratings, color = "b", alpha = 0.1)
plt.xlabel("Cocoa Percent")
```
